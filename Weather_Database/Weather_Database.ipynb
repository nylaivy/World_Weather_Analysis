```python
import pandas as pd
import random
import time
import numpy as np
from citipy import citipy
# Import the requests library.
import requests

# Import the API key.
from config import weather_api_key
```

```python
x = np.random.uniform(-90.000, 90.000, size=2000)
y = np.random.uniform(-90.000, 90.000, size=2000)
coordinates = zip (x, y)
```

```python
city_name = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # If the city is unique, then we will add it to the cities list.
    if city not in city_name:
        city_name.append(city)

len(city_name)
```

```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
weather_data = []

# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1

# Loop through all the cities in the list.
for i, city in enumerate(city_name):

    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")
    
    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Add 1 to the record count.
    record_count += 1
    
    try:

        # Make a 'Get' request for the city weather. ADD SLEEP/PAUSE FUNCTION FOR API RULES
        city_weather = requests.get(city_url)
        time.sleep(1)


        #Get the text from the Get request and convert to json
        city_data = city_weather.json()

        #Retreive information from the data
        lat = city_data["coord"]["lat"]
        lng = city_data["coord"]["lon"]
        max_temp = city_data["main"]["temp_max"]
        humidity = city_data["main"]["humidity"]
        clouds = city_data["clouds"]["all"]
        wind = city_data["wind"]["speed"]
        description = city_data["weather"][0]["description"]
        country = city_data["sys"]["country"]

        #Store the data in a dataframe
        weather_data.append({"City": city.title(),
                             "Country": country,
                             "Lat": lat,
                             "Lng": lng,
                             "Max Temp": max_temp,
                             "Humidity": humidity,
                             "Cloudiness": clouds,
                             "Wind Speed": wind,
                             "Current Description": description})
         
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
weather_data_df = pd.DataFrame(weather_data)
weather_data_df.head()
```

```python
weather_data_df.to_csv(r'WeatherPy_Database.csv')
```
